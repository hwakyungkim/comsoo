```python
## 전처리

myrawdata = rawdata.copy()
data_document = myrawdata['답안']
Answer_Num = data_document.count()
target_data = myrawdata[pnum2]

if process == '기본':
  for i, text in enumerate(data_document):
    text = text.lower()
    data_document[i] = kiwi.space(text)

elif pnum == '1번':
  kiwi.add_user_word('삼각형', 'NNP', 10)
  kiwi.add_user_word('오각형', 'NNP', 10)
  kiwi.add_user_word('정오각형', 'NNP', 11)
  kiwi.add_user_word('선분', 'NNP', 10)

  for i, text in enumerate(data_document):
    text = text.lower()
    text = text.replace("변", "선분")
    text = text.replace("모르게따", "모르겠다")
    text = text.replace("센치", "cm")
    text = text.replace("상각", "삼각")
    text = text.replace("크키", "크기")
    text = text.replace("모르갰음", "모르겠다")
    text = text.replace("성분", "선분")
    text = text.replace("산분", "선분")
    text = text.replace("떄문", "때문")
    text = text.replace("떄뭉네", "때문에")
    text = text.replace("비씨쥐", "bcg")
    text = text.replace("비쥐", "bg")
    text = text.replace("에이 비 에프", "abf")
    text = text.replace("에이 에프", "af")
    text = text.replace("비쥐", "bg")
    #text = text.replace("gb", "bg") # 이거 나중에 추가한거임
    text = text.replace("꼭지점", "꼭짓점")
    text = text.replace("정 오각형", "정오각형")
    data_document[i] = kiwi.space(text)

elif pnum == '2번':
  kiwi.add_user_word('에서', 'NNP', 10)
  kiwi.add_user_word('분속', 'NNP', 10)
  kiwi.add_user_word('구간', 'NNP', 10)

  for i, text in enumerate(data_document):
    text = text.lower()
    text = text.replace("이동거리", "이동 거리")
    text = text.replace("키로", "km")
    text = text.replace("킬로", "km")
    text = text.replace("속도", "속력")
    text = text.replace("키로미터", "km")
    text = text.replace("킬로미터", "km")
    text = text.replace("~", "에서")
    text = text.replace("나무꿍", "나무꾼")
    text = text.replace("도망갇", "도망가")
    text = text.replace("못햤댜", "못했다")
    data_document[i] = kiwi.space(text)
```

```python
# 벡터화

if vector == 'bow':
  from sklearn.feature_extraction.text import CountVectorizer
  if pnum == '2번':
    mindf = 0.003
    if pnum2 == '채점2':
      mindf = 0.004
    elif pnum2 == '채점4':
      mindf = 0.01

  if process == '기본':
    cvect = CountVectorizer()
  elif process == '전문가':
    if pnum == '2번':
      cvect = CountVectorizer(min_df = mindf)
    else:
      cvect = CountVectorizer(min_df = 0.003)
  if dic == 'exam_answer': ## 여기서예시답안으로 fit
    cvect.fit([newanswer])
  else:
    cvect.fit([newanswer])
  train_feature = cvect.transform(train_input)
  test_feature = cvect.transform(test_input)

elif vector == 'bi':
  from sklearn.feature_extraction.text import CountVectorizer
  cvect = CountVectorizer(ngram_range=(2,2), min_df=10) # 2gram
  cvect.fit(train_input)
  train_feature = cvect.transform(train_input)
  test_feature = cvect.transform(test_input)

elif vector == 'tf':
  from sklearn.feature_extraction.text import TfidfVectorizer
  cvect = TfidfVectorizer(min_df=5)
  cvect.fit(train_input)
  train_feature = cvect.transform(train_input)
  test_feature = cvect.transform(test_input)
```
